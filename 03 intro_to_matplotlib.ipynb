```python
#LINEPLOT
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.plot(x,x**2)
ax.set(title="LINE PLOT",
       xlabel="numbers",
       ylabel="squares")
```

```python
#SCATTERPLOT
fig, ax=plt.subplots()
ax.scatter(x,np.sin(x))
ax.set(title="SCATTER PLOT",
       xlabel="numbers",
       ylabel="sine")
```

```python
#BAR PLOT
fig, ax=plt.subplots()
ax.bar(x,np.cos(x))
ax.set(title="BAR PLOT",
       xlabel="numbers",
       ylabel="cosine")
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist()
```

```python
heart_disease.plot.hist(figsize=(10,20),subplots=True)
```

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
#ax.set_xlim((40,100))
```

```python
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'],
           y=over_50['chol'],
           c=over_50['target'])
ax.set(title="OVER 50 HEART DISEASE",
       xlabel="AGE",
       ylabel="CHOLESTROL")
ax.legend(*scatter.legend_elements(),title="target")
ax.axhline(over_50["chol"].mean(),
           linestyle="--")
```

```python
over_50
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                          sharex=True)

scat1=ax1.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"])
scat2=ax2.scatter(x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"])
ax1.set(title="CHOLESTROL V/S AGE",
      #  xlabel="AGE",
        ylabel="CHOLESTROL")
ax2.set(title="HEART RATE V/S AGE",
        xlabel="AGE",
        ylabel="HEART RATE")
ax1.legend(*scat1.legend_elements(),title="TARGET A")
ax2.legend(*scat2.legend_elements(),title="TARGET B")
ax1.axhline(over_50["chol"].mean(),
            linestyle="--")
ax2.axhline(over_50["thalach"].mean(),
            linestyle="--")
fig.suptitle("HEART DISEASE PRACTICE",fontsize=20,fontweight="bold")
```

```python
plt.style.use("seaborn-whitegrid")
fig,(ax1,ax2)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                          sharex=True)

scat1=ax1.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"],
            cmap="plasma")
ax1.set_xlim([50,80])
scat2=ax2.scatter(x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"],
            cmap="summer")
ax2.set_xlim([50,80])
ax2.set_ylim([60,200])
ax1.set(title="CHOLESTROL V/S AGE",
      #  xlabel="AGE",
        ylabel="CHOLESTROL")
ax2.set(title="HEART RATE V/S AGE",
        xlabel="AGE",
        ylabel="HEART RATE")
ax1.legend(*scat1.legend_elements(),title="TARGET A")
ax2.legend(*scat2.legend_elements(),title="TARGET B")
ax1.axhline(over_50["chol"].mean(),
            linestyle="--")
ax2.axhline(over_50["thalach"].mean(),
            linestyle="--")
fig.suptitle("HEART DISEASE PRACTICE",fontsize=20,fontweight="bold")
```
