```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1, random_state=37,hypercube=False,class_sep=30)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(X[:,0],X[:,1],c=y,s=80)
```

```python
np.random.seed(3)
def percetron(X,y,lr,ep):
    X=np.insert(X,0,1,axis=1)
    weights = np.random.rand(X.shape[1])
    learning_rate = lr
    epochs = ep
    
    for i in range(epochs):
        index = np.random.randint(0,1000)
        sum_of_weights = np.dot(X[index],weights) 
        result = activation_function(sum_of_weights)
        weights = weights + learning_rate*(y[index]-result)*X[index]
        
    slope = -(weights[1]/weights[2])
    intercept = -(weights[0]/weights[2])
    
        
    return weights[0], weights[1:]

def activation_function(val):
    if(val<=0):
        return 0
    else:
        return val
```

```python
bias, weights = percetron(X,y,0.1,1000)
```

```python
print(bias)
print(weights)
```

```python
slope = -(weights[0]/weights[1])
intercept = -(bias/weights[1])
```
