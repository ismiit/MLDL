```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization
```

```python
data = pd.read_csv('data/digit-recognizer/train.csv')
data.head()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
y_train.shape
```

```python
y_train = y_train.values
y_test = y_test.values
```

```python
ohe = OneHotEncoder(sparse=False,handle_unknown='ignore')
y_train = ohe.fit_transform(y_train.reshape(-1,1))
y_test = ohe.transform(y_test.reshape(-1,1))
```

```python
y_train.shape
```

```python
def evaluation(model,X,y):
    evala = model.evaluate(X,y)
    loss = evala[0]
    accuracy = evala[1]
    print('loss = ',np.round(loss,5))
    print('accuracy = ',np.round(accuracy*100,2))
```

# Sigmoid 
 * sigmoid(x) = 1 / (1 + exp(-x))
 * 0 < sigmoid(x) < 1

```python
sig = Sequential()
```

```python
sig.add(Dense(16,activation='sigmoid'))
sig.add(Dense(16,activation='sigmoid'))
sig.add(Dense(10,activation='sigmoid'))
```

```python
sig.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
eps = 100
model_sig = sig.fit(X_train,y_train,epochs=eps,validation_split=0.2,validation_data=(X_test,y_test))
```

```python
evaluation(sig,X_test,y_test)
```

# (ELU) Exponential Linear Unit

* elu(x) = x  , if x > 0 
* elu(x) = alpha * (exp(x) - 1)  , if x < 0.

```python
elu = Sequential()
```

```python
elu.add(Dense(16,activation='elu'))
elu.add(Dense(16,activation='elu'))
elu.add(Dense(10,activation='elu'))
```

```python
elu.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
eps = 100
model_elu = elu.fit(X_train,y_train,epochs=eps,validation_split=0.2,validation_data=(X_test,y_test))
```

```python
evaluation(elu,X_test,y_test)
```

# SOFTMAX 
* softmax(x) = exp(x) / tf.reduce_sum(exp(x))

```python
soft = Sequential()
```

```python
soft.add(Dense(16,activation=tf.keras.activations.softmax))
soft.add(Dense(16,activation=tf.keras.activations.softmax))
soft.add(Dense(10,activation='sigmoid'))
```

```python
soft.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
eps = 100
model_soft = soft.fit(X_train,y_train,epochs=eps,validation_split=0.2,validation_data=(X_test,y_test))
```

```python
evaluation(soft,X_test,y_test)
```
