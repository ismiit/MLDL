```python
X=car_sales_missing.drop("Price",axis=1)
y=car_sales_missing["Price"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
cat_imputer=SimpleImputer(strategy="constant",fill_value="missing")
door_imputer=SimpleImputer(strategy="constant",fill_value=4)
num_imputer=SimpleImputer(strategy="mean")
cat_features=["Make","Colour"]
door_features=["Doors"]
num_features=["Odometer (KM)"]
transformer=ColumnTransformer([
    ("cat_imputer",cat_imputer,cat_features),
    ("door_imputer",door_imputer,door_features),
    ("num_imputer",num_imputer,num_features)
])
filled_X=transformer.fit_transform(X)
car_sales_filled=pd.DataFrame(filled_X,
                             columns=["Make","Colour","Doors","Odometer (KM)"])
one_hot=OneHotEncoder()
transformer_new=ColumnTransformer([("one-hot",
                                  one_hot,
                                  categorical_features)],
                                  remainder="passthrough")
trans_X=transformer_new.fit_transform(car_sales_filled)
trans_X
```

```python
np.random.seed(2)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test=train_test_split(trans_X,
                                                y,
                                                test_size=0.2)
```

```python
model=RandomForestRegressor()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
heart_dis=pd.read_csv("data\heart-disease.csv")
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
np.random.seed(2)
X=heart_dis.drop("target",axis=1)
y=heart_dis["target"]
X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2)
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
clf_single_score=clf.score(X_test,y_test) #score method in classifier gives the mean accuracy of the model
clf_accuracy=np.mean(cross_val_score(clf,X,y,cv=5))
```

```python
print(f"Classfier Cross Validated Accuracy: {clf_accuracy*100:.2f}%")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
house_prices=pd.read_csv("data/HousingData.csv")
house_prices["CRIM"].fillna(house_prices["CRIM"].mean(),inplace=True)
house_prices["ZN"].fillna(house_prices["ZN"].mean(),inplace=True)
house_prices["INDUS"].fillna(house_prices["INDUS"].mean(),inplace=True)
house_prices["CHAS"].fillna(house_prices["CHAS"].mean(),inplace=True)
house_prices["AGE"].fillna(house_prices["AGE"].mean(),inplace=True)
house_prices["LSTAT"].fillna(house_prices["LSTAT"].mean(),inplace=True)
np.random.seed(2)
X=house_prices.drop("MEDV",axis=1)
y=house_prices["MEDV"]
X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2)
model=RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)   #in regressor score method gives the coefficeint of determination (R2 score)
reg_single_score=model.score(X_test,y_test)
reg_cross_val_score=np.mean(cross_val_score(model,X,y,cv=5))
reg_single_score, reg_cross_val_score
```
