```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y  = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
plt.figure(figsize=(14,7))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1
    
    for i in range(1000):
        index = np.random.randint(0,100)
        y_hat = result(np.dot(X[index],weights))
        weights = weights + (learning_rate * (y[index] - y_hat) * X[index])
        
    return weights[0], weights[1:]

def result(val):
    return 1 if (val > 0) else 0 
```

```python
intercept , slopes = perceptron(X,y)
```

```python
print(intercept)
print(slopes)
```

```python
m = -(slopes[0] / slopes[1])
b = -(intercept / slopes[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+b
```

```python
plt.figure(figsize=(14,7))
plt.plot(x_input,y_input,color='red')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

## How perceptron different from Logistic Regression ?

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X,y)
```

```python
clf.coef_ , clf.intercept_
```

```python
A = clf.coef_[0][0]
B = clf.coef_[0][1]
C = clf.intercept_[0]
m2 = -(A / B)
b2 = -(C / B)
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m2*x_input1+b2
```

```python
plt.figure(figsize=(14,7))
plt.plot(x_input,y_input,color='red')
plt.plot(x_input1,y_input1,color='black')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
